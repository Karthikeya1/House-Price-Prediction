Training Multiple Models

```python
#Linear, Ridge, Lasso,  SVR, Adaboost, Gradient boost, XGboost, KNN, Decision Tree, Random Forest
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.pipeline import Pipeline
```

```python
#Train test splitting the data and proceeding with out scaling
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
'''
Tree based models like decision tree,random forest,gradient boost,XGboost,Adaboost doesn't need scaled or standardized data
Distance based models like knn,kmeans and gradient based models like Linear,Ridge,Lasso needs scaled data
With out scaling the data, split it into training and testing
For scaling required models, use pipeline method in scikit-learn to apply preprocessing and model in same step
'''

models={
    'Linear Regression':Pipeline([
                                ("scalar",StandardScaler()),
                                ("model",LinearRegression())
                                ]),
    'Ridge':Pipeline([
                    ("scalar",StandardScaler()),
                    ("model",Ridge(alpha=1.0))
                    ]),
    'Lasso':Pipeline([
                    ("scalar",StandardScaler()),
                    ("model",Lasso(alpha=0.1))
                    ]),
    'Elastic Net':Pipeline([
                            ("scalar",StandardScaler()),
                            ("model",ElasticNet(alpha=1.0))
                            ]),
    'Support Vector Regressor':Pipeline([
                                ("scalar",StandardScaler()),
                                ("model",SVR(C=100,gamma=0.1,kernel="rbf"))
                                ]),
    'KNN':Pipeline([
                    ("scalar",StandardScaler()),
                    ("model",KNeighborsRegressor(n_neighbors=5))
                    ]),
    'DecisionTree':DecisionTreeRegressor(max_depth=10),
    'Random Forest':RandomForestRegressor(n_estimators=100),
    'Adaboost':AdaBoostRegressor(n_estimators=100),
    'Gradient Boost':GradientBoostingRegressor(n_estimators=100),
    'XG boost':XGBRegressor(n_estimators=100,learning_rate=0.1)
}
```

```python
#Function for evaluating the model
def evaluate_model(y_true,y_pred):
    from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error
    mse=mean_squared_error(y_true,y_pred)
    mae=mean_absolute_error(y_true,y_pred)
    rmse=root_mean_squared_error(y_true,y_pred)
    score=r2_score(y_true,y_pred)
    return mse,mae,rmse,score
```

```python
#Model Building, Training and Evaluation
for name,model in models.items():
    mo=model

    #Training the model
    mo.fit(X_train,y_train)

    #Evaluating the model
    y_train_pred=mo.predict(X_train)
    mse1,mae1,rmse1,score1=evaluate_model(y_train,y_train_pred)
    y_test_pred=mo.predict(X_test)
    mse2,mae2,rmse2,score2=evaluate_model(y_test,y_test_pred)

    #Printing the model metrics
    print("="*30)
    print("Model : "+name)
    print("-"*30)
    print("Training Data :")
    print("Accuracy : ",score1)
    print("Mean Squared Error : ",mse1)
    print("Mean Absolute Error : ",mae1)
    print("Root Mean Squared Error : ",rmse1)
    print("-"*30)
    print("Testing Data :")
    print("Accuracy : ",score2)
    print("Mean Squared Error : ",mse2)
    print("Mean Absolute Error : ",mae2)
    print("Root Mean Squared Error : ",rmse2)
    print("="*30)
    print("\n\n\n")

    
```

The Best performing models are Random Forest, Gradient Boost and XG Boost.
Applying Hyperparameter tuning for the best performing models to get better accuracy.

Hyperparameter Tuning

```python
rf_params = {"max_depth": [5, 8, 15, None, 10],
             "max_features": [5, 7, "auto", 8],
             "min_samples_split": [2, 8, 15, 20],
             "n_estimators": [100, 200, 500, 1000]}

gradient_params={"loss": ['squared_error','huber','absolute_error'],
             "criterion": ['friedman_mse','squared_error','mse'],
             "min_samples_split": [2, 8, 15, 20],
             "n_estimators": [100, 200, 500],
              "max_depth": [5, 8, 15, None, 10],
            }

xgboost_params = {"learning_rate": [0.1, 0.01],
                  "max_depth": [5, 8, 12, 20, 30],
                  "n_estimators": [100, 200, 300],
                  "colsample_bytree": [0.5, 0.8, 1, 0.3, 0.4]}
```

```python
grid_search_models=[
    ("Random Forest",RandomForestRegressor(),rf_params),
    ("Gradient Boosting",GradientBoostingRegressor(),gradient_params),
    ("XG Boosting",XGBRegressor(),xgboost_params)
]
```

```python
import warnings
warnings.filterwarnings("ignore")


from sklearn.model_selection import RandomizedSearchCV


best_params={}
for name,model,params in grid_search_models:
    model=RandomizedSearchCV(estimator=model,
                             param_distributions=params,
                             cv=5,
                             n_iter=100,
                             n_jobs=-1,
                             verbose=2)

    model.fit(X_train,y_train)

    best_params[name]=model.best_params_


for model in best_params:
    print(f"------------------Best params for {model}--------------------------------")
    print(best_params[model])
```

```python
model.feature_importances_
```

```python
#Training the models using the best parameter
model1=RandomForestRegressor(n_estimators=100,min_samples_split=8,max_features=8)
model2=GradientBoostingRegressor(n_estimators=200,min_samples_split=8,max_depth=5,loss="huber",criterion="squared_error")
model3=XGBRegressor(n_estimators=300,max_depth=5,learning_rate=0.1,colsample_bytree=0.5)

models={
    "Random Forest" : model1,
    "Gradient Boosting" : model2,
    "XG Boosting" : model3
}
```

```python
#Model Building, Training and Evaluation
for name,model in models.items():
    mo=model

    #Training the model
    mo.fit(X_train,y_train)

    #Evaluating the model
    y_train_pred=mo.predict(X_train)
    mse1,mae1,rmse1,score1=evaluate_model(y_train,y_train_pred)
    y_test_pred=mo.predict(X_test)
    mse2,mae2,rmse2,score2=evaluate_model(y_test,y_test_pred)

    #Printing the model metrics
    print("="*30)
    print("Model : "+name)
    print("-"*30)
    print("Training Data :")
    print("Accuracy : ",score1)
    print("Mean Squared Error : ",mse1)
    print("Mean Absolute Error : ",mae1)
    print("Root Mean Squared Error : ",rmse1)
    print("-"*30)
    print("Testing Data :")
    print("Accuracy : ",score2)
    print("Mean Squared Error : ",mse2)
    print("Mean Absolute Error : ",mae2)
    print("Root Mean Squared Error : ",rmse2)
    print("="*30)
    print("\n\n\n")

    
```

Therefore, Gradient Boosting is the best algorithm for this dataset.
